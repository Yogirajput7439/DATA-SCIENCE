```python
import matplotlib.pyplot as plt
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
labels = ["sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
explode = [0.1, 0, 0, 0]
colors = [ "red", "green", "black", "yellow"]
plt.title("Career runs for Indians Batsmen")
plt.pie(runs, labels=labels, autopct='%1.1f%%', explode=explode, colors=colors, shadow=True, startangle=360)
plt.show()
```

```python
labels = ["Samsung", "Apple", "Xiomi", "Vivo", "Oppo", "Others"]
sales = [600, 487, 407, 286, 250, 933]

colors = ["green", "yellow", "red", "pink", "blue", "orange"]

plt.title("Mobile Sales in 2024")
plt.pie(sales, labels=labels, autopct='%1.1f%%', colors=colors)
plt.show()
```

```python
labels = ["Samsung", "Apple", "Xiomi", "Vivo", "Oppo", "Others"]
sales = [600, 487, 407, 286, 250, 933]
explode = [0.1, 0, 0, 0, 0, 0]
colors = ["green", "yellow", "red", "pink", "blue", "orange"]

plt.title("Mobile Sales in 2024")
plt.pie(sales, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode)
plt.show()
```
