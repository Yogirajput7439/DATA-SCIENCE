```python
# now we will see class

class Company:
    def __init__(self, name):
        self.name = name
        
    def info(self):
        
        print(f"the boy who is best in coding and their journey is {self.name}")

c1 = Company("yogesh")
c1.info()
```

```python

class College:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def illustrate(self):

        print(f"my name is {self.name} and my age is {self.age}")

s1 = College("yogesh", 21)
s2 = College("sham", 21)

s1.illustrate()
s2.illustrate()
```
