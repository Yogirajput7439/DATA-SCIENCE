```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data
data = {
    'Age': [20, 25, 30, 35, 40],
    'Salary': [20000, 30000, 40000, 50000, 60000]
}
df = pd.DataFrame(data)

# Step 2: Standardisation
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled_data, columns=['Age', 'Salary'])

# Step 3: Plot before scaling
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.scatterplot(x='Age', y='Salary', data=df, marker='o', s=100, color='blue')
plt.title("Original Data")
plt.xlabel("Age")
plt.ylabel("Salary")

# Step 4: Plot after scaling
plt.subplot(1,2,2)
sns.scatterplot(x='Age', y='Salary', data=df_scaled, marker='o', s=100, color='red')
plt.title("Standardised Data (mean=0, std=1)")
plt.xlabel("Age (scaled)")
plt.ylabel("Salary (scaled)")

plt.tight_layout()
plt.show()
```
