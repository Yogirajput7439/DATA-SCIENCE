```python
df = df[df["Pclass"].isin([1,2,3])]
```

```python
sns.countplot(data=df, x=df['Pclass'])
```

```python
pclass_count = df['Pclass'].value_counts().sort_index()
plt.pie(
    pclass_count,
    labels = pclass_count.index,
    autopct = "1.1f%%",
    startangle=90
)

plt.title("passanger distribution by class")
plt.show()
```
