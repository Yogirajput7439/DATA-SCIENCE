```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Data (y = 2x)
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

# Gradient Descent Parameters
w = 0
lr = 0.01
history = []

# Run gradient descent and record weight/loss
for _ in range(100):
    y_pred = w * x
    error = y - y_pred
    loss = (error ** 2).mean()
    dw = -2 * (x * error).mean()
    w = w - lr * dw
    history.append((w, loss))

# Extract weights and losses for animation
weights = [h[0] for h in history]
losses = [h[1] for h in history]

# Set up the figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 2.5)
ax.set_ylim(0, max(losses))
line, = ax.plot([], [], 'r-', lw=2)
point, = ax.plot([], [], 'bo')

# Initialize function
def init():
    line.set_data([], [])
    point.set_data([], [])
    return line, point

# Animation function
def update(frame):
    line.set_data(weights[:frame], losses[:frame])
    point.set_data(weights[frame-1], losses[frame-1])
    return line, point

# Create animation
ani = FuncAnimation(fig, update, frames=len(weights), init_func=init,
                    blit=True, interval=100, repeat=False)

# Save animation as GIF
gif_path = "/mnt/data/gradient_descent_animation.gif"
ani.save(gif_path, writer='pillow')

gif_path
```
