```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Folder to save graphs
os.makedirs("gradient_descent_steps", exist_ok=True)

# Data: y = 2x
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

# Initial weight and learning rate
w = 0
lr = 0.01
weights = []
losses = []

# Run gradient descent and save plots at intervals
for step in range(1, 101):
    y_pred = w * x
    error = y - y_pred
    loss = (error ** 2).mean()
    dw = -2 * (x * error).mean()
    w = w - lr * dw

    weights.append(w)
    losses.append(loss)

    # Save graph every 10 steps
    if step % 10 == 0:
        plt.figure()
        plt.plot(weights, losses, 'r-')
        plt.scatter(weights[-1], losses[-1], color='blue')
        plt.title(f"Step {step}: Weight vs Loss")
        plt.xlabel("Weight (w)")
        plt.ylabel("Loss")
        plt.grid(True)
        plt.savefig(f"gradient_descent_steps/step_{step}.png")
        plt.close()

print("✅ Static graphs saved in 'gradient_descent_steps' folder")
```
