```python
from sklearn import tree
```

```python
features = [
    [150, 0],
    [170, 0],
    [130, 1],
    [120, 1],
]

labels = ["apple", "apple", "orange", "oranges"]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(features, labels)

print(clf.predict([[150, 0]]))  # Heavy and red-ish? Probably an apple!
print(clf.predict([[115, 1]])) 
```

```python
from sklearn import tree

# Step 1: Collect training data
features = [[140, 1], [130, 1], [150, 0], [170, 0]]  # [weight, texture]
labels = [0, 0, 1, 1]  # 0 = Apple, 1 = Orange

# Step 2: Train the model
clf = tree.DecisionTreeClassifier()
clf = clf.fit(features, labels)

# Step 3: Make a prediction
print(clf.predict([[190, 0]]))  # Predicts: Orange (1)
```
